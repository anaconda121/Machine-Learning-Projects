```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
i = misc.ascent() #getting image from scipy library
```

```python
#graphing original image
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(i)
plt.show()
```

```python
i_transformed = np.copy(i) #cloning array i
print(i_transformed) 
size_x = i_transformed.shape[0]
print(size_x)
size_y = i_transformed.shape[1]
print(size_y)
```

```python
# This filter detects edges nicely
# It creates a convolution that only passes through sharp edges and straight
# lines.

# filter for detecting vertical edges
filter = [ 
    [-1, -2, -1], 
    [0, 0, 0], 
    [1, 2, 1]
]

'''
#filter for detecting horizontal edges
filter = [ 
    [-1, 0, 1], 
    [-2, 0, 2], 
    [-1, 0, 1]
]
'''
```

```python
#convolution operation from scratch
#x and y start at 1 because you can't perform a convolution on index 0, its on the edge
for x in range(1,size_x-1):
    #nesting the for loops because the nested loop needs to iterate over all the y-coords per x-coord
    for y in range(1,size_y-1):
        #applying filter to image
        convolution = 0.0
        convolution += (i[x - 1, y-1] * filter[0][0])
        convolution += (i[x, y-1] * filter[0][1])
        convolution += (i[x + 1, y-1] * filter[0][2])
        convolution += (i[x-1, y] * filter[1][0])
        convolution += (i[x, y] * filter[1][1])
        convolution += (i[x+1, y] * filter[1][2])
        convolution += (i[x-1, y+1] * filter[2][0])
        convolution += (i[x, y+1] * filter[2][1])
        convolution += (i[x+1, y+1] * filter[2][2])
        
        #error handlers for exceptions
        if(convolution < 0):
            convolution = 0
        if(convolution > 255):
            convolution = 255
        
        #replacing original value of x and y with the convolution var
        i_transformed[x, y] = convolution
```

```python
# Plot the image. Note the size of the axes -- they are 512 by 512
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
plt.show()      
```

```python
#pooling algorithm - 2 by 2 pooling filter
new_x = int(size_x/2)
new_y = int(size_y/2)
newImage = np.zeros((new_x, new_y))
#third variable is 2 because the stride of the pooling algorithm is 2
for x in range(0, size_x, 2):
    for y in range(0, size_y, 2):
        pixels = []
        pixels.append(i_transformed[x - 1, y])
        pixels.append(i_transformed[x, y])
        pixels.append(i_transformed[x, y - 1])
        pixels.append(i_transformed[x - 1, y - 1])
        newImage[int(x/2),int(y/2)] = max(pixels)

# Plot the image. Note the size of the axes -- now 256 pixels instead of 512
plt.gray()
plt.grid(False)
plt.imshow(newImage)
plt.show()      
```
